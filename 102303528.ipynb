```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
r = 102303528

ar = 0.05 * (r % 7)
br = 0.3 * (r % 5 + 1)

print(f"Constants calculated: ar = {ar:.4f}, br = {br:.4f}")

try:
    df = pd.read_csv('/content/data.csv', encoding='latin1', low_memory=False)
except UnicodeDecodeError:
    df = pd.read_csv('/content/data.csv', encoding='ISO-8859-1', low_memory=False)

x = df['no2'].dropna().values

z = x + ar * np.sin(br * x)

def pdf_model(z, c, lam, mu):
    return c * np.exp(-lam * (z - mu)**2)

counts, bin_edges = np.histogram(z, bins=100, density=True)
z_bins = (bin_edges[:-1] + bin_edges[1:]) / 2

initial_guess = [np.max(counts), 0.1, np.mean(z)]

params, covariance = curve_fit(pdf_model, z_bins, counts, p0=initial_guess)

c_learned, lambda_learned, mu_learned = params

print("\nüîπ Learned Parameters for Submission:")
print(f"‚û°Ô∏è c      : {c_learned:.6f}")
print(f"‚û°Ô∏è lambda : {lambda_learned:.6f}")
print(f"‚û°Ô∏è mu     : {mu_learned:.6f}")
```
